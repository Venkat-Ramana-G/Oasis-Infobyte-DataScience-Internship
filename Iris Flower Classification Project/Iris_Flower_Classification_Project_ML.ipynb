**Model Training:-**

Importing all necessary libraries for modeling

```python
# Splitting the data using SKlearn Module
from sklearn.model_selection import train_test_split 

# importing warnings and ignoring them
import warnings
warnings.filterwarnings('ignore')

X = df.drop(columns=['species'])
Y = df['species']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25,random_state = 32)
```

Calculating the accuracy score and the prediction by using Logistic Regression:-

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# model training
model.fit(x_train, y_train)
```

```python
# print metric to get performance
print("Accuracy: ",model.score(x_test, y_test) * 100)
```

Calculating the accuracy score and the prediction by using K-nearest neighbours:-

```python
# knn - k-nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()

model.fit(x_train, y_train)

print("Accuracy: ",model.score(x_test, y_test) * 100)
```

Calculating the accuracy score and the prediction by using Decision Tree:-

```python
# decision tree
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

model.fit(x_train, y_train)

print("Accuracy: ",model.score(x_test, y_test) * 100)

```

**Pickling the model:-**

```python
# save the model
import pickle
filename = 'iris_model.pkl'
pickle.dump(model, open(filename, 'wb'))
```
